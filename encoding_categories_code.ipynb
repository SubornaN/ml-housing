# Code for encoding to ordinal values

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data_selected_columns.csv') # loading csv
```

```python
pd.options.display.max_columns  = 60 # showing max rows
data.head(10) # showing the first 10 rows
```

```python
data.columns # list of all columns
```

```python
categories = ['MSZoning', 'LotShape', 'LotConfig', 'Neighborhood', 'Condition1', 'BldgType', 'HouseStyle', 'YearBuilt', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'MasVnrType', \
             'ExterQual', 'ExterCond',  'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', \
             'Functional', 'FireplaceQu', 'GarageType', 'GarageYrBlt',  'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'MiscFeature', 'SaleType', 'SaleCondition']
```

```python
from sklearn.preprocessing import LabelEncoder # importing encoder library

label = LabelEncoder() # creating an instance of the class LabelEncoder
# looping through every column and changing the values into ordinal
for column in categories:
    data[column] = label.fit_transform(data[column])    
    
```

```python
data.to_csv('encoded_selected_columns.csv', index = False) # saving the dataframe as csv
```
