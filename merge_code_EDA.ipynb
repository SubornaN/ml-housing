```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('utitilies_house_struc/House_Stucture_ Utilities_EDA.csv')
df1 = df1.rename(columns = {'Id': 'id'})
df1.head()
```

```python
df2 = pd.read_csv('bsmt_and_misc/bsmt_and_misc3.csv')
df2 = df2.rename(columns = {'Id': 'id'})
df2.head()
```

```python
df3 = pd.read_csv('garage_house_qualities/garage.final.csv')
df3 = df3.drop(['Unnamed: 0', 'garagetype_NA', 'garagefinish_NA'], axis=1)
df3 = df3.fillna(0)
df3.head()
```

```python
df4 = pd.read_csv('garage_house_qualities/house_qualities.final.csv')
df4 = df4.drop('Unnamed: 0', axis=1)
df4.head()
```

```python
df5 = pd.read_csv('land_ext/land_ext.csv')
df5 = df5.drop('Unnamed: 0', axis=1)
df5.head()
```

```python
df_merge_col = pd.merge(df1, df2, on='id')
df_merge_col = pd.merge(df_merge_col, df3, on='id')
df_merge_col = pd.merge(df_merge_col, df4, on='id')
df_merge_col = pd.merge(df_merge_col, df5, on='id')
df_merge_col
```

```python
df_merge_col.to_csv('Merged_df.csv', index = False)
```
